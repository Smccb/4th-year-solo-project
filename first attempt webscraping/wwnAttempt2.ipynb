```python
import requests

url = 'https://waterfordwhispersnews.com/'
response = requests.get(url)

if response.status_code == 200:
    main_page_content = response.text
else:
    print(f"Failed to fetch the main page. Status code: {response.status_code}")
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(main_page_content, 'html.parser')

# If the links are directly in <a> tags
article_links = soup.select('a')


# Now 'article_links' contains a list of links to individual articles
for link in article_links:
    print(link.get('href'))
```

```python
for link in article_links:
    article_url = link['href']  # Replace with the correct attribute containing the URL
    article_response = requests.get(article_url)

    if article_response.status_code == 200:
        article_content = article_response.text
        
        # Now you can extract the headline using BeautifulSoup and the h1 tag
        article_soup = BeautifulSoup(article_content, 'html.parser')
        headline = article_soup.find('h1', {'itemprop': 'headline'}).text
        print(headline)
    else:
        print(f"Failed to fetch article page. Status code: {article_response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
from xml.etree import ElementTree as ET

url = 'https://waterfordwhispersnews.com/wp-sitemap-posts-post-1.xml'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:94.0) Gecko/20100101 Firefox/94.0'
}

sitemap_response = requests.get(url, headers=headers)

if sitemap_response.status_code == 200:
    sitemap_content = sitemap_response.text

    # Parse the XML sitemap
    root = ET.fromstring(sitemap_content)

    # Iterate through the URLs in the sitemap
    for url_elem in root.iter('{http://www.sitemaps.org/schemas/sitemap/0.9}url'):
        loc_elem = url_elem.find('{http://www.sitemaps.org/schemas/sitemap/0.9}loc')
        article_url = loc_elem.text

        # Fetch the content of each article URL
        article_response = requests.get(article_url, headers=headers)

        if article_response.status_code == 200:
            article_content = article_response.text

            # Now you can extract the headline using BeautifulSoup and the appropriate HTML tags
            article_soup = BeautifulSoup(article_content, 'html.parser')
            
            # Replace 'your_headline_selector' with the correct selector for the headline
            headline = article_soup.find('h1', {'itemprop': 'headline'}).text
            print(headline)
        else:
            print(f"Failed to fetch article page. Status code: {article_response.status_code}")

else:
    print(f"Failed to fetch sitemap. Status code: {sitemap_response.status_code}")
```

```python
import requests

def check_robots_txt(url):
    try:
        robots_txt_url = f"{url.rstrip('/')}/robots.txt"
        response = requests.get(robots_txt_url)

        if response.status_code == 200:
            print("robots.txt content:")
            print(response.text)
        else:
            print(f"Failed to fetch robots.txt. Status code: {response.status_code}")

    except Exception as e:
        print(f"Error: {str(e)}")

# Example usage:
check_robots_txt("https://waterfordwhispersnews.com/")
```
