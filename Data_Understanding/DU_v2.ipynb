1. Separate the dataset into two categories based on the "is_sarcastic" column.

2. Tokenize the text in each category.

3. Calculate word and token statistics for each category.

```python
import pandas as pd
from collections import Counter
from nltk.tokenize import word_tokenize



file_path = 'Train_v1_cleaned_without_column.csv'

# Read the dataset
data = pd.read_csv(file_path)


# Separate dataset into sarcastic and non-sarcastic categories
sarcastic_texts = data[data['is_sarcastic'] == 1]['text']
non_sarcastic_texts = data[data['is_sarcastic'] == 0]['text']

# Function to tokenize text and calculate word and token statistics
def calculate_statistics(texts):
    # Tokenize texts
    tokens = [word_tokenize(text.lower()) for text in texts]
    
    # Flatten tokens list
    tokens_flat = [token for sublist in tokens for token in sublist]
    
    # Count unique words and total tokens
    unique_words = len(set(tokens_flat))
    total_tokens = len(tokens_flat)
    
    # Calculate word frequency
    word_freq = Counter(tokens_flat)
    
    return unique_words, total_tokens, word_freq

# Calculate statistics for sarcastic texts
sarcastic_unique_words, sarcastic_total_tokens, sarcastic_word_freq = calculate_statistics(sarcastic_texts)

# Calculate statistics for non-sarcastic texts
non_sarcastic_unique_words, non_sarcastic_total_tokens, non_sarcastic_word_freq = calculate_statistics(non_sarcastic_texts)

# Print statistics
print("Sarcastic Texts:")
print("Unique words:", sarcastic_unique_words)
print("Total tokens:", sarcastic_total_tokens)
print("Word frequency:", sarcastic_word_freq)
print("\nNon-sarcastic Texts:")
print("Unique words:", non_sarcastic_unique_words)
print("Total tokens:", non_sarcastic_total_tokens)
print("Word frequency:", non_sarcastic_word_freq)
```
