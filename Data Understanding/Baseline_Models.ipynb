```python
import numpy as np
from sklearn.svm import SVC

file_path = 'Train_v1.txt'
column_names = ['toRemove', 'is_sarcastic', 'text']

# Read the dataset
data = pd.read_csv(file_path, sep='\t', header=None, names=column_names)

shuffled_data = data.sample(frac=1, random_state=1).reset_index(drop=True)

# Define the number of chunks
num_chunks = 5

# Split the data into chunks
data_chunks = np.array_split(shuffled_data, num_chunks)

# Iterate over each chunk
for i, chunk in enumerate(data_chunks):
    print(f"Processing chunk {i+1}/{num_chunks}")
    
    # Build count matrix for the chunk
    count_matrix_chunk = tmu.build_count_matrix(chunk.text)
    
    # Train and evaluate the classifier
    clf = SVC(random_state=1)
    y_chunk = chunk.is_sarcastic
    tmu.printClassifReport(clf, count_matrix_chunk, y_chunk)
```

```python
import numpy as np
import text_mining_utils as tmu


# Define the size of the sample
sample_size = 10000

# Sample from the dataset
sampled_data = shuffled_data.sample(n=sample_size, random_state=1)

total_rows = len(sampled_data)
print(total_rows)
```

```python
class_distribution = sampled_data['is_sarcastic'].value_counts()
print(class_distribution)
```

```python
baseline_count_matrix = tmu.build_count_matrix(sampled_data.text)
baseline_count_matrix.head()
```

```python
baseline_tf_matrix = tmu.build_tf_matrix(sampled_data.text)
baseline_tf_matrix.head()
```

```python
baseline_tfidf_matrix = tmu.build_tfidf_matrix(sampled_data.text)
baseline_tfidf_matrix.head()
```

```python
#baseline model preformace
from sklearn.svm import SVC

clf = SVC(random_state=1)
y = sampled_data.is_sarcastic

tmu.printClassifReport(clf, baseline_count_matrix, y)
tmu.printClassifReport(clf, baseline_tf_matrix, y)
tmu.printClassifReport(clf, baseline_tfidf_matrix, y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(random_state=1)

y = sampled_data.is_sarcastic

tmu.printClassifReport(dt_clf, baseline_count_matrix, y)
tmu.printClassifReport(dt_clf, baseline_tf_matrix, y)
tmu.printClassifReport(dt_clf, baseline_tfidf_matrix, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_clf = KNeighborsClassifier()

y = sampled_data.is_sarcastic

tmu.printClassifReport(KNN_clf, baseline_count_matrix, y)
tmu.printClassifReport(KNN_clf, baseline_tf_matrix, y)
tmu.printClassifReport(KNN_clf, baseline_tfidf_matrix, y)
```
